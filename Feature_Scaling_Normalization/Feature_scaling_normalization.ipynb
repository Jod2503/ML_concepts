```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('wine_data.csv',header=None,usecols=[0,1,2])
df.columns=['Class_label','alcohol','malic_acid']
df
```

Normalization is the process to change the values of numeric columns to use a column scale
1. Min-max scaling:
   
   

The final distribution lies between [0,1] 



The geometrical inference is fitting a 3d curve into a cube and 2d data into a square

```python
sns.distplot(x=df['alcohol'],hist=False)
```

```python
sns.distplot(x=df['malic_acid'],hist=False,color='red')
```

```python
sns.scatterplot(x=df['alcohol'],y=df['malic_acid'],hue=df['Class_label'])
```

```python
# perform train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop('Class_label', axis=1),
                                                    df['Class_label'],
                                                    test_size=0.3,
                                                    random_state=0)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()

scaler.fit(X_train)

X_train_scaled=scaler.transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

```python
X_train_scaled
```

```python
# convert into data frame
X_train_scaled=pd.DataFrame(X_train_scaled,columns=X_train.columns)
X_test_scaled=pd.DataFrame(X_test_scaled,columns=X_test.columns)
```

```python
X_train_scaled
```

```python
np.round(X_train_scaled.describe(),1)
```

```python
# scatterplot before and after scaling
sns.scatterplot(x=X_train['alcohol'],y=X_train['malic_acid'],color='yellow')
plt.title('Before Scaling')
plt.show()
sns.scatterplot(x=X_train_scaled['alcohol'],y=X_train_scaled['malic_acid'],color='red')
plt.title('after scaling')
plt.show()
```

```python
sns.distplot(X_train['alcohol'],hist=False,label='alcohol')
sns.distplot(X_train['malic_acid'],hist=False,label='Malic_acid')
plt.title('Before Scaling')
plt.legend()
plt.show()
sns.distplot(X_train_scaled['alcohol'],hist=False,label='alcohol')
sns.distplot(X_train_scaled['malic_acid'],hist=False,label='Malic_acid')
plt.title('After Scaling')
plt.legend()
plt.show()
```
